```python
df = pd.read_csv('Data/ratebeer_dataset_all.csv')
df
```

```python
no_null = df['city'].dropna(axis=0)
no_null
```

```python
cities = no_null.unique()
```

```python
cities
```

```python
len(cities)
```

```python
locs = []
x = 0;
for c in cities:
    locs.append(geolocator.geocode(c, timeout=4))
    print(x)
    sleep(1)
    x += 1
```

```python
new_dict = {}
for i in range(len(cities)):
    if(locs[i] is None):
        new_dict[cities[i]] = (None, None)
    else:
        new_dict[cities[i]] = (locs[i].latitude, locs[i].longitude)
```

```python
new_dict
```

```python
new_dict['Sukhum'][1]
```

```python
df['lat'] = [new_dict.get(c, (None, None))[0] for c in df['city']]
df['lon'] = [new_dict.get(c, (None, None))[1] for c in df['city']]
```

```python
df.to_csv('Data/ratebeer_dataset_all_with_loc.csv')
```
