```python
from geopy.geocoders import Nominatim
import pandas as pd
from time import sleep
```

```python
geolocator = Nominatim()
```

```python
location = geolocator.geocode("St. Petersburg, Russia")
```

```python
print((location.latitude, location.longitude))
```

```python
df = pd.read_csv('data/ratebeer_dataset_all.csv')
df[df['country'] == 'Russia']
```

```python
no_null = df[['city', 'country']].dropna(axis=0)
no_null
```

```python
concat_name = no_null['city'] + ', ' + no_null['country']
```

```python
concat_name
```

```python
cities = concat_name.unique()
```

```python
cities
```

```python
len(cities)
```

```python
locs = []
x = 0;
for c in cities:
    locs.append(geolocator.geocode(c, timeout=4))
    print(x)
    sleep(1)
    x += 1
```

```python
new_dict = {}
for i in range(len(cities)):
    if(locs[i] is None):
        new_dict[cities[i]] = (None, None)
    else:
        new_dict[cities[i]] = (locs[i].latitude, locs[i].longitude)
```

```python
new_dict
```

```python
new_dict['Sukhum, Abkhazia'][1]
```

```python
df['lat'] = [new_dict.get(c, (None, None))[0] for c in df['city'] + ', ' + df['country']]
df['lon'] = [new_dict.get(c, (None, None))[1] for c in df['city'] + ', ' + df['country']]
```

```python
df.to_csv('data/ratebeer_dataset_all_with_loc.csv')
```
